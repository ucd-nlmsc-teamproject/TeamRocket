## Data Collection from the API

```python
import numpy as np
import pandas as pd
import requests
```

```python
def Countries_Global_Summary(url):
    response = requests.get(url)
    data = response.json()
    df = pd.DataFrame(data['Countries'])
    df2 = pd.DataFrame(data['Global'],index=[0])
    Countries = list(df["Slug"])
    Countries.sort()
    return df, df2, Countries
```

```python
Data_Countries, Data_Global, Countries = Countries_Global_Summary("https://api.covid19api.com/summary")
```

```python
def Data_collection(Countries):
    Data = pd.DataFrame()
    for i in range(len(Countries)):
        A = Countries[i]
        df = pd.read_json('https://api.covid19api.com/total/country/%s'%A)
        Data = Data.append(df)
    
    return Data
```

```python
Data = Data_collection(Countries)
print(Data)
```

```python
Data.to_csv("Raw_Data.csv",index=False)
```
