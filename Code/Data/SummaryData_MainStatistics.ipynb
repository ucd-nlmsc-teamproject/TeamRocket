```python
#Fetching data from covid-19 API by Johns Hopkinns

import pandas as pd
import numpy as np
import urllib.request
import os
import csv
import requests
import json
```

```python
link = 'https://api.covid19api.com/summary' #summary data link
```

```python
response = requests.get(link) #using the get method here
```

```python
covid_data = response.json()
```

```python
covid_data['Countries'] #using the values in the dictionary key: Countries
```

```python
df = pd.DataFrame(covid_data['Countries'])
df
```

### Above is the dataframe that contains our main data that we'd refer for statistical references!

```python
df = df.reindex(sorted(df.columns), axis = 1) #re-arranging the columns
```
