```python
def Lastest_date_data(df_old,df_new):
    df_latest = pd.DataFrame()
    T = df_old.iloc[[-1]].index.tolist()[0]
    Today = datetime.today().strftime('%Y-%m-%d')
    d1 = datetime.strptime(T, "%Y-%m-%d")
    d2 = datetime.strptime(Today, "%Y-%m-%d")
    Diff = abs((d2 - d1).days)
    for i in range(1,Diff):
        df_latest = df_latest.append(df_new.iloc[[-i]])
    
    df_latest = df_latest.sort_index()
    
    return df_latest
    
```
