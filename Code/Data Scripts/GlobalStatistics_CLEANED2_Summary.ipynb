### This is the CLEANED Summary Data for 248 countries of all the required Statistics!

#### It contains the following values for 248 countries
#### New Confirmed, Total Confirmed
#### New Deaths, Total Deaths
#### New Recovered, Total Recovered

```python
#Fetching data from covid-19 API by Johns Hopkinns

import pandas as pd
import numpy as np
import urllib.request
import os
import csv
import requests
import json
```

```python
def get_data(link):               #defining a function to get covid19 data
    response = requests.get(link) #using the get method here
    covid_data = response.json()  #extracting the data
    df = pd.DataFrame(covid_data['Countries']) #creating a df from the data above and selecting the Key: Countries
    df = df.reindex(sorted(df.columns), axis = 1) #re-arranging the columns for enhanced readability
    df = df.drop(columns = ['Slug','Premium')  #dropping the column SLUG and Premium
    df['Date'] = pd.to_datetime(df['Date'])
    df['Date'] = df['Date'].dt.date
    return df
    
```

```python
link = 'https://api.covid19api.com/summary'
covid_summary_df = get_data(link)
```

```python
covid_summary_df.tail()
```

```python
#Above we can see a total of 186 countries
#Function to get the details for the other countries which have no cases recorded
#In all this dataframe would now contain 248 countries

def Adding_null_country(df):
    df_country = pd.read_json('https://api.covid19api.com/countries')  #reading the file for All Country names
    Con_all = df_country['Country']  #storing all 248 country names
    Country_1 = df['Country']  #Passing already present countries
    filt_list = list(set(Con_all) - set(Country_1))  #Subtracting the two lists to get the one's we need
    iso_list =[]  #country code
    con_name =[]  
    for i in range(len(filt_list)): #Using for loop to add additional countries and information for those countries
        df_country2 = df_country.loc[df_country.Country == filt_list[i]]   
        con_name = con_name + list(df_country2['Country'])
        iso_list = iso_list + list(df_country2['ISO2'])
    Data_f = pd.DataFrame(columns = df.columns)
    Data_f['Country'] = con_name
    Data_f['CountryCode']= iso_list
    Data_f['Date'] = df['Date']
    Data_f.fillna(0, inplace = True)  #filling the null values, placing 0
    Sum_tot = pd.concat([df, Data_f], ignore_index = True, sort = False)
    return Sum_tot
```

```python
covid_global_summary = Adding_null_country(covid_summary_df)
```

```python
covid_global_summary
```

```python
#As we can see this has 248 countries in all
covid_global_summary = covid_global_summary.sort_values(by=['Country'], ascending = True)
```
